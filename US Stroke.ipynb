```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("Stroke_Mortality_Data_Among_US_Adults__35___by_State_Territory_and_County___2015-2017.csv")
# df  = df.loc[df['Year'] == 2016]
print(df.size)
df = df.loc[ df['Data_Value'].notnull()]
df = df.loc[ df['Stratification2'] == "Overall"]
df = df.loc[ df['GeographicLevel'] ==  "County"]

print(df.size)
```

```python
df
```

```python
def show(dataFrame, minValue):
    dataFrame['text'] = dataFrame['LocationDesc'] + '<br>' + dataFrame['Data_Value'].astype(str)
    dataFrame = dataFrame.loc[dataFrame['Data_Value'] > minValue ]
    
    fig = go.Figure(data=go.Scattergeo(
            lon = dataFrame['X_lon'],
            lat = dataFrame['Y_lat'],
            text = dataFrame['text'],
            mode = 'markers',
            marker = dict(
                size = 10,
                opacity = 0.8,
                reversescale = False,
                autocolorscale = False,
                symbol = 'circle',
                line = dict(
                    width=1,
                    color='rgba(102, 102, 102)'
                ),
                colorscale = 'rainbow',
                cmin = minValue,
                color = dataFrame['Data_Value'],
                cmax = dataFrame['Data_Value'].max(),
                colorbar_title="per 100,000 population"
            )
        ))

    fig.update_layout(
            title = 'Stroke Instances in USA, 2015-2017',
            geo_scope='usa',
        )
    fig.show()
```

```python
show(df.copy(), 0)
show(df.copy(), 50)
show(df.copy(), 100)
show(df.copy(), 150)
```

```python
import time

def convert_loc_to_coordinates(df):
    start = time.time()
    y_lats = []
    x_longs = []
    n=0
    for loc in df2['Location 1']:
        y_lat = float(loc.split(',')[0][1:])
        x_long = float(df2['Location 1'][0].split(',')[1][1:-1])
        y_lats.append(y_lat)
        x_longs.append(x_longs)
        print(str(time.time()-start) + "seconds")
        n+=1
        print(n)
    df['Y_lat'] = y_lats
    df['X_lon'] = x_longs
    return df
```
